1. Resize an image using scaling factors

```python
import cv2
from google.colab.patches import cv2_imshow
from google.colab import files

uploaded = files.upload()
img = cv2.imread(list(uploaded.keys())[0])

resized = cv2.resize(img, None, fx=0.5, fy=0.5)

cv2_imshow(resized)
```

2. Rotate an image by a given angle

```python
import cv2
from google.colab.patches import cv2_imshow
from google.colab import files

uploaded = files.upload()
img = cv2.imread(list(uploaded.keys())[0])

(h, w) = img.shape[:2]
M = cv2.getRotationMatrix2D((w//2, h//2), 45, 1)

rotated = cv2.warpAffine(img, M, (w, h))

cv2_imshow(rotated)
```

3. Crop a specific portion of an image

```python
import cv2
from google.colab.patches import cv2_imshow
from google.colab import files

uploaded = files.upload()
img = cv2.imread(list(uploaded.keys())[0])

crop = img[50:200, 50:200]

cv2_imshow(crop)
```

4. Flip image horizontally and vertically

```python
import cv2
from google.colab.patches import cv2_imshow
from google.colab import files

uploaded = files.upload()
img = cv2.imread(list(uploaded.keys())[0])

h_flip = cv2.flip(img, 1)
v_flip = cv2.flip(img, 0)

cv2_imshow(h_flip)
cv2_imshow(v_flip)
```

5. Adjust brightness and contrast

```python
import cv2
from google.colab.patches import cv2_imshow
from google.colab import files

uploaded = files.upload()
img = cv2.imread(list(uploaded.keys())[0])

bright = cv2.convertScaleAbs(img, alpha=1.5, beta=50)

cv2_imshow(bright)
```
